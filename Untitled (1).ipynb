### 1.1. Análisis Faltantes

Nuestro primer paso en el análisis exploratorio será estudiar la existencia de datos nulos y en su caso tratarlos.

```python
df.isna().any()
```

Observamos que la gran mayoría de variables presentan datos nulos.
De entre de entre las variables: 'Neighbourhood', 'City', 'State','Zipcode', 'Market'y'Smart Location', podemos eliminar todas salvo una pues nos aportan una información muy similar. Dado que la única que no presenta datos nulos es 'Smart Location' nos quedaremos con ella.

```python
df = df.drop(['Neighbourhood', 'City', 'State','Zipcode', 'Market'], axis=1)
```

Además eliminamos la variable 'Country' por ser linealmente dependiente de 'Country Code' y presentra faltantes.

```python
df = df.drop(['Country'], axis=1)
```

De entre el resto de variables que presentan datos nulos eliminamos además 'First Review' puesto que no es relevante cuando se hizo la primera review, sí lo es cuando se hizo la última. También eliminamos 'Jurisdiction Names' y 'Reviews per Month'.

```python
df = df.drop([ 'First Review', 'Jurisdiction Names', 'Reviews per Month'], axis=1)
```

```python
df.isna().any()
```

Observamos que la lista de variables que presentan datos nulos se ha reducido, sin embargo; aún hay una gran cantidad de variables con missings. Para acabar con este problema usaremos la clase Imputer de sklearn y susutituiremos los datos nulos por la mediana. Este procedimiento solo sirve para las variables que no son objects, estás últimas las trataremos individualmente.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
```

```python
df_num = df.select_dtypes(exclude=['object'])
df_num.info()
```

```python
imputer.fit(df_num)
```

```python
df_num_nonan = imputer.transform(df_num)
```

```python
df_ = pd.DataFrame(df_num_nonan, columns=df_num.columns)
```

```python
df[df_num.columns]=df_
df.isna().any()
```

Observamos que hemos conseguido eliminar los datos missing de las variables numéricas pero aún quedan los del resto. Vamos a tratar de eliminarlos también.

```python
df_cat_na=df[['Amenities','Has Availability', 'Last Review', 'License','Features']]
df_cat_na.info()
```

Observamos que las variables 'License' y 'Has Availability' son en casi todas las observaciones nulas. Por tanto, las eliminamos. Por otro lado la variable 'Last Review' presenta bastantes menos datos missing que las anteriores pero un gran número, aún así. Por ello también la eliminamos. Por último para acabar definitivamente con los missings eliminamos las filas correspondientes.

```python
df=df.drop(['License','Has Availability','Last Review'], axis=1)
```

```python
df=df.dropna()
```

```python
df.isna().any()
```
